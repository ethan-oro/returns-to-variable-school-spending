```python
def grab_data(url):
    page_content = None
    try: 
        page = requests.get(url)
        if page.status_code == 200:
            page_content = BeautifulSoup(page.content, 'html.parser')
        else:
            return None
    except Exception as error:
        print('Caught this error while loading page content (get_page_content()): ' + repr(error))
        return None
    block = page_content.findAll('script')[8]
    block = str(block)
    a = re.findall('\[.*]', block)
    code = url.split('/')[5].split('-')[0]
    desired = None
    for b in a[1:]:
        try:
            if (b[b.index('\'')+1 : b.index('\'', b.index('\'') + 1)] == code):
                desired.append(b.replace('\'', '').replace('[', '').split(','))
        except ValueError:
            continue
    return desired
```

```python

"""
    Instructions: 
    1. to import, type: ' from afdb_scraper_downloads_only import * ' 
    2. to run, type: ' run() '
    The downloaded documents will be located ' ./../../data/pdf_downloads/afdb-scraper/ ' 
""" 

complaints_index_url = 'https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2907172/'

def get_page_content(url):
    """
        Given a url, this would return the html content of the page parsed by BeautifulSoup
    """
    try: 
        page = requests.get(url)
        page_content = BeautifulSoup(page.content, 'html.parser')
        return page_content
    except Exception as error:
        print('Caught this error while loading page content (get_page_content()): ' + repr(error))
        return

    
def get_project_table(html):
    """ 
        Given the html content for the projects index, returns dataframe of project table, 
        that includes project reference number and project name. 
    """
    maindiv = html.find('div', {'id': 'c136818'})
    table = maindiv.find_all('table')
    df = pd.read_html(str(table), header=0)
    return df[0]


def get_project_urls(html):
    """
        Retrieve the urls and returns a string of the urls, separated by commas.
    """
    urls = []
    all_tr = html.find('tbody').findAll('tr')
    for tr in all_tr: 
        href_match = re.findall('<a href=".*?">', str(tr))[0]
        base_url = 'https://www.afdb.org'
        project_url = base_url+href_match.split('"')[-2]
        urls.append(project_url)
    return ','.join(urls)

def get_download_links(html): 
    """ 
        Given the html content of the project page, returns a string
        of the project downloadable documents separated by comma.
    """
    base_url = 'https://www.afdb.org'
    download_links = []
    download_tag = html.findAll('a', {'class': 'download pdf'})
    for element in download_tag: 
        download_url = base_url + element['href']
        download_links.append(download_url)
    return ','.join(download_links)



def scrape(): 
    """ 
        Scrapes all complaints from afdb webpage. Returns dataframe 
        of complaint information.
    """
    
    html = get_page_content(complaints_index_url)
    df = get_project_table(html)
    
    urls = get_project_urls(html).split(',')
    df['HYPERLINK'] = urls
    
    for x in range(df.shape[0]):
        complaint_id = df.iat[x,0]
        project_url = df.iat[x,2]
        project_page = get_page_content(project_url)
        project_documents = get_download_links(project_page)
        
        if project_documents != '': 
            prepare_download('afdb-scraper', complaint_id, project_documents)

    return

def run(): 
    """ 
        does-all function that scrapes relevant download documents
        data and downloads data. Does not include CSV.
    """
    scrape()
    return
```
